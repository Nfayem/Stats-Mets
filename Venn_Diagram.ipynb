# VENN DIAGRAM ANALYSIS

```python
"""imports for required libraries"""

import pandas as pd
import pandas as pd
import venn 
import warnings
warnings.filterwarnings('ignore')



"""Imoport for dataframe to be analysed"""

df = pd.read_csv('9.20manualFIC.csv')



"""Venn Diagram for Full Dataset"""

Hodf = df.loc[(df['HOMA'] >= 3.0)]
Fdf = df.loc[(df['FBS'] >= 100)]
Hbdf = df.loc[(df['HbA1C%'] >= 8.8)]

labels = venn.generate_petal_labels([
      set(Fdf['ID'].to_list()), 
      set(Hodf['ID'].to_list()),
      set(Hbdf['ID'].to_list()),  
       ])
fig, ax = venn.venn3(labels, names=['FBS', 'HOMA', 'HbA1C%',])
# fig.show()

"""Insert preferred image name in parenthesis in line 34 below
and uncomment it to save the output as png file for export"""

# fig.savefig("VD_full")



"""Venn diagrams based on age ranges"""

below_ten =  df.loc[(df['AGE'].isin(range(0,9,1)))]
ten_sixteen = df.loc[ (df['AGE'].isin(range(10,16,1)))]
adult = df.loc[ (df['AGE'].isin(range(16,18,1)))]

def VD_Age(data):
    
      Hodf = data.loc[(df['HOMA'] >= 3.0)]
      Fdf = data.loc[(df['FBS'] >= 100)]
      Hbdf = data.loc[(df['HbA1C%'] >= 8.8)]

      labels = venn.generate_petal_labels([
      set(Fdf['ID'].to_list()), 
      set(Hodf['ID'].to_list()),
      set(Hbdf['ID'].to_list()),
  
       ])
      fig, ax = venn.venn3(labels, names=['FBS', 'HOMA', 'HbA1C%',])
      # fig.show()

      """Insert preferred image name in parenthesis in line 62 below
      and uncomment it to save the output as png file for export"""

      fig.savefig("VD_adult")
      
VD_Age(adult)
```
